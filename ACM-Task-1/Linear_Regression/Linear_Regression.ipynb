## Import Libraries: NumPy, Pandas and matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read dataset and get YOE and Salary data

```python
data = pd.read_csv("Salary_dataset.csv", index_col=0)

yoe = np.array([yoe for yoe in data['YearsExperience']])
salaries = np.array([salary for salary in data['Salary']])
```

## Scatter Plot of YOE v/s Salary

```python
plt.scatter(yoe, salaries, c='r',marker='x')
plt.title("Salary v/s YOE Comparision")
plt.xlabel('$YOE$')
plt.ylabel('$Salary$')
plt.show()
```

## Finding w, b and using Gradient Descent

```python
w = 0
b = 0
dj_dw = 0
dj_db = 0
num_iterations = 100000


def pred_func(x):
  return w*x + b


def gradient_descent(x_train, y_train, a):

  global w, b, dj_dw, dj_db

  temp_dj_dw = 0
  temp_dj_db = 0
  m = x_train.shape[0]

  for i in range(m):

    temp_dj_dw += ((pred_func(x_train[i]) - y_train[i])*x_train[i])/m
    temp_dj_db += (pred_func(x_train[i]) - y_train[i])/m

  dj_dw = temp_dj_dw
  dj_db = temp_dj_db

  w = w - a*(dj_dw)
  b = b - a*(dj_db)


for i in range(num_iterations):
  gradient_descent(yoe, salaries, 0.01)

print(f"w value: {w}")
print(f"bias value: {b}")
```

## Plot linear regression line w.r.t scatter plot

```python
plt.scatter(yoe, salaries, c='r', marker='x')
plt.title("Salary v/s YOE Comparision")
plt.xlabel('$YOE$')
plt.ylabel('$Salary$')

x = np.linspace(0, 12, 1000)
y = w*x + b
plt.plot(x, y, c='b', label="Linear Reg line")
plt.legend()

plt.show()
```
